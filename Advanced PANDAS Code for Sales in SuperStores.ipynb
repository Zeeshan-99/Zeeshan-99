# Which sub-Categories has Standard Deviation for Sales and Profit one point more than the Mean of Sales and Profit?

```python
for i in salesData.columns:
#     print(len(np.unique(salesData[i])))
      print('The Total unique no of ', i ,'=', len(np.unique(salesData[i])))
```

```python
salesData.columns
```

```python
meanInfo= pd.pivot_table(salesData,values=['Sales','Profit']\
               ,index=['Sub-Category']\
               ,aggfunc='mean')\
               .reset_index()

meanInfo.columns=['Sub-Category','MeanSales','MeanProfit']
meanInfo
# list(salesData.columns)
```

```python
stdInfo= pd.pivot_table(salesData\
                       ,index=['Sub-Category']\
                       ,values=['Sales','Profit']\
                       , aggfunc='std')\
                        .reset_index()


stdInfo.columns= ['Sub-Category','stdSales','stdProfit']
stdInfo
```

```python
savTheTable=pd.merge(meanInfo, stdInfo, on='Sub-Category', how='left')
```

```python
savTheTable.head(2)
```

```python
savTheTable['diffMean']=savTheTable['MeanProfit']-savTheTable['stdProfit']
```

```python
savTheTable['diffStd']=savTheTable['MeanSales']-savTheTable['stdSales']
```

```python
savTheTable.sort_values(by='diffStd', ascending=False)
```

```python
from abc import abstractmethod,ABC

class TouchScreenLaptop(ABC):


    @abstractmethod

    def scroll(self):

        pass


    @abstractmethod

    def click(self):

        pass


class HP(TouchScreenLaptop):


    def scroll(self):

        super().scroll()

        print("Touchpad")


class Dell(TouchScreenLaptop):


    def scroll(self):

        super().scroll()

        print("Mouse")


class HPNotebook(HP):

    def click(self):

        print("touchpad mouse")


class DellNotebook(Dell):

    def click(self):

        print("USB mouse")


h=HPNotebook()

h.click()

h.scroll()

d=DellNotebook()

d.scroll()

d.click()
```
